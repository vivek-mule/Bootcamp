```python
import pandas as pd

ds = pd.read_csv("Sales.csv")

ds
```

```python
ds["Revenue"] = ds["Price"] * ds["QuantitySold"]
ds
```

```python
mean = ds["Price"].mean()
std = ds["Price"].std()

z_scores = (ds["Price"] - mean) / std

ds["Z Score"] = z_scores

ds
```

```python
ds = ds[abs(z_scores) < 2]
ds
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()

ds[["Price", "Revenue"]] = scaler.fit_transform(ds[["Price", "Revenue"]])
ds
```

```python
ds = ds.drop("Z Score", axis = 1)
ds
```
