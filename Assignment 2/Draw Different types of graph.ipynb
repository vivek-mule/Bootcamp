```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds = pd.read_csv("Housing.csv")

ds
```

```python
plt.plot(ds["price"], ds["area"])
plt.xlabel("Price")
plt.ylabel("Area")

plt.title("Line Plot")
plt.show()
```

```python
plt.plot(ds["bedrooms"], ds["price"], color = "green")
plt.xlabel("Bedrooms")
plt.ylabel("Price")

plt.title("Line Plot")
plt.show()
```

```python
plt.plot(ds["stories"], ds["price"], color = "orange")
plt.xlabel("Stories")
plt.ylabel("Price")

plt.title("Line Plot")
plt.show()
```

```python
plt.plot(ds["furnishingstatus"], ds["price"], color="black")
plt.xlabel("Furnishing Status")
plt.ylabel("Price")

plt.title("Line Plot")
plt.show()
```

```python
plt.bar(ds["area"], ds["price"], color = "red")
plt.xlabel("Area")
plt.ylabel("Price")

plt.show()
```

```python
plt.bar(ds["bedrooms"], ds["price"], color = "blue")

plt.xlabel("Bedrooms")
plt.ylabel("Price")

plt.show()
```

```python
plt.scatter(ds["price"], ds["area"], color = "pink")
plt.xlabel("Price")
plt.ylabel("Area")

plt.show()
```

```python
sns.boxplot(x="bedrooms", y="price", data=ds)

plt.title("Bedroom VS Price")
plt.xlabel("Number of Bedrooms")
plt.ylabel("Price")

plt.show()
```

```python
sns.barplot(x="furnishingstatus", y="price", data=ds, hue="furnishingstatus")
```

```python
pivot_table = ds.pivot_table(index="bathrooms", columns="stories", values="price", aggfunc="mean")

sns.heatmap(pivot_table, annot=True, cmap="rocket", linewidth=0.5)

plt.title("Average Price : Bathrooms VS Stories")
plt.xlabel("Number of stories")
plt.ylabel("Number of Bathrooms")
plt.show()
```

```python
sns.countplot(x="furnishingstatus", hue="mainroad", data=ds)

plt.title("Number of Properties by Furnishing Status and Main Road")
plt.xlabel("Furnishing Status")
plt.ylabel("Count")

plt.show()
```

```python
sns.histplot(x="price", data=ds, hue="parking", palette="deep")

plt.title("Distribution of Property Prices")
plt.xlabel("Price")
plt.ylabel("Count")
```
